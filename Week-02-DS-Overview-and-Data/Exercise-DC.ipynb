### Clean the Data

Perhaps we’d like to know more another day, but the team is really interested in just the relationship between calories (`calories_day`) and weight. …and maybe gender.

But what about the weird non-numeric values in the column obviously meant for numeric data?

### Bonus #2

Filter out the non-single-states (e.g., `'California, Colorado'`) in the most elegant way possible (i.e., *not* by blacklisting all the bad values).

```python
# Filter out rows where 'state' contains a comma, indicating multiple states
single_state_df = df_salary[df_salary['state'].apply(lambda x: isinstance(x, str) and ',' not in x)]

# Group by state and calculate the minimum, mean, and maximum salary
single_state_salary_stats = single_state_df.groupby('state')['salary'].agg(['min', 'mean', 'max'])

# Sort by mean salary in descending order
single_state_salary_stats_sorted = single_state_salary_stats.sort_values(by='mean', ascending=False)

single_state_salary_stats_sorted.head()
```

### Bonus #3

Show the quantiles instead of just minimum, mean, and maximum―say 0%, 5%, 25%, 50%, 75%, 95%, and 100%. Outliers may be deceiving.

Sort by whatever interests you―like say the *50th* percentile.

And throw in a count by state too. It would be interesting to know how many data points contribute to the figures for each state. (*Hint: your nice formatting from Bonus #1 might not work this time around.* 😜)

```python
# Calculate quantiles and count by state
quantiles = [0, 0.05, 0.25, 0.5, 0.75, 0.95, 1]

# Ensure 'salary' column is numeric
single_state_df.loc[:, 'salary'] = pd.to_numeric(single_state_df['salary'], errors='coerce')

# Calculate quantiles
state_salary_quantiles = single_state_df.groupby('state')['salary'].apply(lambda x: x.quantile(quantiles)).unstack()

# Calculate count
state_salary_count = single_state_df.groupby('state')['salary'].count()

# Combine quantiles and count into a single DataFrame
state_salary_stats = pd.concat([state_salary_quantiles, state_salary_count], axis=1)
state_salary_stats.columns = [f'{int(q*100)}%' for q in quantiles] + ['count']

# Sort by the 50th percentile (median) in descending order
state_salary_stats_sorted = state_salary_stats.sort_values(by='50%', ascending=False)

state_salary_stats_sorted.head()
```
